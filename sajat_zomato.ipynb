```python
import pandas as pd
import os
def load_data(path):
 csv_path = os.path.join(path, "zomato.csv")
 return pd.read_csv(csv_path)

zomato = load_data("datasets")
zomato
```

```python
zomato = zomato.drop(['url', 'name', 'phone', 'location', 'rest_type', 'dish_liked', 'reviews_list', 'menu_item', 'cuisines', 'address'], axis=1)
```

```python
import hashlib
import numpy as np

def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
 ids = data[id_column]
 in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
 return data.loc[~in_test_set], data.loc[in_test_set]

def test_set_check(identifier, test_ratio, hash):
 return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

zomato_with_id = zomato.reset_index()
train_set, test_set = split_train_test_by_id(zomato_with_id, 0.2, "index")
```

```python
from sklearn.preprocessing import LabelEncoder

def yesNoConverter(column):
    encoder = LabelEncoder()
    column_converted = encoder.fit_transform(column)
    return column_converted
```

```python
import string

 
def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False

def cleanVotes(data):
    if(isinstance(data, float)):
        return data
    else:
        cleanData = data.translate({ord(c): None for c in string.whitespace})
        vote = cleanData.split('/')
        if (isfloat(vote[0])):
            return float(vote[0])
        else:
            return np.nan
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hotTransformer(zomato_encoded):
    encoder = OneHotEncoder(categories='auto')
    zomato_cat_1hot = encoder.fit_transform(zomato_encoded.reshape(-1,1))
    return zomato_cat_1hot
```

```python
from sklearn.preprocessing import LabelEncoder

def labelEncoder(column):
    encoder = LabelEncoder()
    zomato_cat_encoded = encoder.fit_transform(column)
    return zomato_cat_encoded
```

```python

train_set['online_order'] = yesNoConverter(train_set['online_order'])
train_set['book_table'] = yesNoConverter(train_set['book_table'])
 
```

```python
train_set_type_encoded = labelEncoder(train_set['listed_in(type)'])
train_set_type_encoded_oneHot = one_hotTransformer(train_set_type_encoded)
train_set_type_encoded_oneHot.toarray()
```

```python
train_set_city_encoded = labelEncoder(train_set['listed_in(type)'])
train_set_city_encoded_oneHot = one_hotTransformer(train_set_city_encoded)
train_set_city_encoded_oneHot.toarray()
```

```python
import statistics

def calculateMedianByColumnName(column):
    items_cleaned = [x for x in column if str(x) != 'nan']
    return statistics.median(items_cleaned)
```

```python
def replaceNanByMedian(column):
    median = calculateMedianByColumnName(column)
    column_cleaned = []
    for x in column:
        if str(x) == 'nan':
            column_cleaned.append(median)
        else:
            column_cleaned.append(x)
    return column_cleaned
```

```python
def standardScaler(column):
    mean = np.mean(column)
    variance = np.var(column)
    standardizedCol = []
    for x in column:
        x = (x - mean) / variance
        standardizedCol.append(x)
   
    return standardizedCol
```

```python
rate_labels = []

for n in train_set['rate']:
    rate_labels.append(cleanVotes(n))
    
zomato_labels = rate_labels
```

```python
rate_labels = replaceNanByMedian(rate_labels)
```

```python
train_set['approx_cost(for two people)'] = replaceNanByMedian(train_set['approx_cost(for two people)'])
```

```python
def Normalize(column):
    min = float(column.min())
    max = float(column.max())
    column_substract = []
    for x in column:
        column_substract.append((float(x)-min)/(max-min))
    return column_substract
```

```python
train_set['votes'] = Normalize(train_set['votes'])
```

```python
def ConvertFromMoneyToString(column):
    column_converted = []
    for x in column:
        column_converted.append(x.replace(",", ""))
    return column_converted
```

```python
train_set['approx_cost(for two people)'] = ConvertFromMoneyToString(train_set['approx_cost(for two people)'])
```

```python
train_set['approx_cost(for two people)'] = Normalize(train_set['approx_cost(for two people)'])
```
